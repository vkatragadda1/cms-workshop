```python
file = os.path.join('water.xyz')
lines = open(file, 'r')
data = lines.readlines()
lines.close()
print(data)
```

```python
import numpy
import os
```

```python
xyz_file = numpy.genfromtxt(fname=file, skip_header=2, dtype='unicode')
```

```python
symbols = xyz_file[:,0]
```

```python
print(symbols)
```

```python
print(xyz_file)
```

```python
coordinates=(xyz_file[:,1:])
```

```python
print(coordinates)
```

```python
coordinates=coordinates.astype(numpy.float)
```

```python
print(coordinates)
```

```python
num_atoms=len(symbols)
print(num_atoms)
```

```python
BLfile = open('bond_lengths.txt', 'w+')
for num1 in range(0, num_atoms):
    for num2 in range(0, num_atoms):
        if num1<num2: 
            x_distance = coordinates[num1,0]-coordinates[num2,0]
            y_distance=coordinates[num1,1]-coordinates[num2,1]
            z_distance=coordinates[num1,2]-coordinates[num2,2]
            bond_length=numpy.sqrt(x_distance**2+y_distance**2+z_distance**2)

            if bond_length>0 and bond_length <= 1.5:
                BLfile.write(F'{symbols[num1]} to {symbols[num2]} : {bond_length:.2f}')
BLfile.close()
```
